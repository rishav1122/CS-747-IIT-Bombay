```python
import numpy as np

file1 = open("outputucb1.txt", "w+")
file2 = open("outputucb2.txt", "w+")

algoname = "ucb-t1"

def algoucb(meanarray,tosstimes,totaltoss):
	ucbarray= meanarray + np.sqrt(2*np.log(totaltoss)/tosstimes)
	bandit = np.argmax(ucbarray)
# 	print((bandit))
	return bandit
```
